```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dff = pd.read_csv(r'C:\Users\User\Pictures\avocado.csv')
```

```python
dff.head()
```

```python
dff = dff.drop('Unnamed: 0', axis=1)
```

```python
dff['Date'] = pd.to_datetime(dff['Date'])
```

```python
dff['Day_of_week'] = dff['Date'].apply(lambda x :x.dayofweek)
dff['Month'] = dff['Date'].apply(lambda x :x.month)
dff['Year'] = dff['Date'].apply(lambda x :x.year)
```

```python
dff = dff.drop('Date', axis=1)
```

```python
dff.head()
```

```python
df1 = pd.get_dummies(dff['type'])
df1
```

```python
df2 = pd.get_dummies(dff['region'])
df2
```

```python
dff = pd.concat([dff, df1, df2], axis=1)
dff.head()
```

```python
dff = dff.drop(['type', 'region'], axis=1)
```

```python
dff.head()
```

```python
X = dff.drop('AveragePrice', axis=1)
y = dff['AveragePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
X.shape
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred, y_test)
```

```python
sns.regplot(y_pred, y_test)
```
